# The Doomsday Algorithm

The algorihm is as of follows:

```python
"""
This is my Python function finding doomsday based on Conway's original Doomsday algorithm
"""
def doomsday(y):
    cent = y // 100
    if y > 1582:
        r = cent % 4
        cent_anchor = [2, 0, 5, 3]
        d = cent_anchor[r]
    else: d = 6 * cent % 7
    t = y % 100
    a = t // 12
    b = t % 12
    c = b // 4
    return (a + b + c + d) % 7
```

# How to Determine Which Year Is Leap?
Pseudocode:
```python
# For Gregorian Calendar
if y > 1582:
    return (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0)
else:
    return y % 4 == 0
# For Revised Julian Calendar (Christmas is 25 Dec but Easter per Julian calendar)
if y > 1923:
    return (y % 4 == 0 and y % 100 != 0) or (y % 900 == 200 or y % 900 == 600)
else:
    return y % 4 == 0
# For Julian Calendar
return y % 4 == 0
```

````python
import os

def clear():
    if os.name == 'nt':
        os.system('cls')
    else: os.system('clear')

def isleap_Greg(y):
    if y > 1582:
        return (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0)
    else:
        return y % 4 == 0
    
def isleap_RJ(y):
    if y > 1923:
        return (y % 4 == 0 and y % 100 != 0) or (y % 900 == 200 or y % 900 == 600)
    else:
        return y % 4 == 0

def isleap_J(y):
    return y % 4 == 0

clear()
y = 2025
if isleap_Greg(y) == isleap_J(y) == isleap_RJ(y):
    print(f"It {'is' if isleap_Greg(y) else 'is not'} a leap year in all calendars")
else:
    if isleap_Greg(y) == isleap_RJ(y):
        print(f"It {'is' if isleap_Greg(y) else 'is not'} a leap year in both Gregorian and Revised Julian calendars")
    else:
        print(f"It {'is' if isleap_Greg(y) else 'is not'} a leap year in Gregorian Calendar")
        print(f"It {'is' if isleap_RJ(y) else 'is not'} a leap year in Revised Julian Calendar")
    print(f"It {'is' if isleap_J(y) else 'is not'} a leap year in Julian Calendar")
````
