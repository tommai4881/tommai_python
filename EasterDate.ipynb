```python
from os import name, system

def clear():
    if name == 'nt':
        system('cls')
    else:
        system('clear')

def doomsday(y):
    return (y + y // 4 - y // 100 - y // 400 - 2) % 7 if y > 1582 else (y + y //4) % 7

def lectionary(y): # In ra chu kỳ bài đọc
    ChuaNhat = ['C', 'A', 'B']
    NgayThuong = ['Chẵn', 'Lẻ']
    return f"Năm {ChuaNhat[y%3]} {NgayThuong[y%2]}"

def delta(y):
    k = y // 100
    return k - k // 4 - 2

def isleap(y):
    return (y % 4 == 0) and (y <= 1582 or y % 100 != 0 or y % 400 == 0)

def NgayTrongTuan(h):
    dow = ["Chúa Nhật", "Thứ Hai", "Thứ Ba", "Thứ Tư", "Thứ Năm", "Jumu'ah", "Sabbath"]
    # Jumu'ah là Thứ Sáu (ngày cầu nguyện Hồi giáo)
    # Sabbath là Thứ Bảy (ngày cầu nguyện đạo Do Thái)
    return dow[h]

class MuaVongvaGiangSinh:
    def __init__(self, year):
        self.year = year
        self.doomsday = doomsday(self.year)
        self.nhuan = isleap(self.year)
    def MuaVong(self): # In ra ngày của Chúa Nhật I Mùa Vọng (vì năm phụng vụ bắt đầu từ Chúa Nhật I Mùa Vọng năm ngoái)
        ChuaNhatMuaVong = [30, 29, 28, 27, 33, 32, 31] if self.nhuan else [29, 28, 27, 33, 32, 31, 30]
        date = ChuaNhatMuaVong[self.doomsday]
        if date > 30: return f"{date-30} Décembre, {self.year - 1}"
        else: return f"{date} Novembre, {self.year - 1}"
    def GiangSinh(self): # In ra ngày trong tuần của Lễ Giáng Sinh và Tết Tây (vì nếu Lễ Giáng Sinh năm ngoái là Thứ Tư thì Tết Dương Lịch cũng ngày Thứ Tư)
        return NgayTrongTuan((doomsday(self.year)-3)%7) if self.nhuan else NgayTrongTuan((doomsday(self.year)-2)%7)
    def ChristmasDate(self):
        return f"{self.GiangSinh()}, 25 Décembre, {self.year - 1}"
    def NewYearDate(self): # In ra ngày của Tết Tây - Lễ Thánh Maria Đức Mẹ Chúa Trời
        return f"{self.GiangSinh()}, 1 Janvier, {self.year}"
    def ThanhGiaThat(self): # In ra ngày của lễ Thánh Gia Thất
        ThanhGia = [28, 27, 26, 25, 31, 30, 29] if self.nhuan else [27, 26, 25, 31, 30, 29, 28]
        date = ThanhGia[self.doomsday]
        if date == 25: return f"Jumu'ah, 30 Décembre, {self.year-1}"
        # Theo Sách Lễ Rôma, Nếu date là 25 (tức là Giáng Sinh rơi vào Chủ Nhật),
        # Lễ Thánh Gia Thất sẽ là Thứ Sáu, ngày 30 tháng 12
        else: return f"Chúa Nhật, {date} Décembre, {self.year-1}"
    def HienLinhVN(self): # In ra ngày Chúa Nhật Hiển Linh
        HienLinh = [4, 3, 2, 8, 7, 6, 5] if self.nhuan else [3, 2, 8, 7, 6, 5, 4]
        return f"{HienLinh[self.doomsday]} Janvier, {self.year}"
    def HienLinhHV(self): # In ra ngày Hiển Linh 6 janvier (cử hành ngoài Việt Nam)
        return f"{NgayTrongTuan((doomsday(self.year)+2)%7)}, 6 Janvier, {self.year}" if self.nhuan else f"{NgayTrongTuan((doomsday(self.year)+3)%7)}, 6 Janvier, {self.year}"
    def CCPRVN(self):
        # Nếu Lễ Hiển Linh ngày 7 hay 8 janvier thì ngày hôm sau là Lễ Chúa Chịu Phép Rửa.
        baptem = [11, 10, 9, 8, 7, 13, 12] if self.nhuan else [10, 9, 8, 7, 13, 12, 11]
        date = baptem[self.doomsday]
        if baptem[self.doomsday] < 9:
            return f"Thứ Hai, {date + 1} Janvier, {self.year}"
        else:
            return f"Chúa Nhật, {date} Janvier, {self.year}"
    def CCPRHV(self):
        baptem = [11, 10, 9, 8, 7, 13, 12] if self.nhuan else [10, 9, 8, 7, 13, 12, 11]
        date = baptem[self.doomsday]
        return f"Chúa Nhật, {date} Janvier, {self.year}"

class LePhucSinh:
    def __init__(self, year):
        self.year = year
    def epact(self):
        g = self.year % 19 + 1
        f = (11 * g) % 30
        if self.year < 1582: return (f - 3) % 30
        else:
            k = self.year // 100
            s = delta(self.year)
            l = 8 * (k - 14) // 25
            return (f - s + l) % 30
    def pfm(self):
        e = self.epact()
        if e not in [24, 25]:
            return 44 - e if e < 24 else 74 - e
        elif e == 24:
            return 49
        elif e == 25:
            return 48 if self.year % 19 > 10 else 48
    def easterday(self):
        m = self.pfm()
        w = (doomsday(self.year) + m) % 7
        return m + 7 - w
    def easterdate(self):
        if self.easterday() > 31:
            return f"{self.easterday()-31} Avril, {self.year}"
        else:
            return f"{self.easterday()} Mars, {self.year}"

class MuaChayvaPhucSinh:
    def __init__(self, year):
        self.year = year
        self._LePhucSinh = LePhucSinh(self.year).easterday()
        self.nhuan = isleap(self.year)
    def ThuongNienTienChay(self):
        interval = 17 if self.nhuan else 18
        return 4 + (self._LePhucSinh - interval) // 7
    def LeTro(self): #Tính ngày của Thứ Tư Lễ Tro
        febdate = 29 if self.nhuan else 28
        ash = self._LePhucSinh - 46
        if ash < 1: return f"{ash+febdate} Février, {self.year}"
        else: return f"{ash} Mars, {self.year}"
    def LePhucSinh(self):
        return LePhucSinh(self.year).easterdate()
    def AscensionThursday(self): # Thứ Năm Thăng Thiên ở Miền Bắc
        ascension = self._LePhucSinh + 39 # Thứ Năm
        if ascension == 61: # Thăng Thiên trùng ngày Giải phóng Miền Nam
            return f"30 Avril, {self.year}"
        elif ascension <= 92:
            return f"{ascension-61} Mai, {self.year}"
        else:
            return f"{ascension-92} Juin, {self.year}"
    def AscensionSunday(self): # Chúa Nhật Thăng Thiên ở Miền Nam
        ascension = self._LePhucSinh + 42 # Chúa Nhật
        if ascension <= 92:
            return f"{ascension-61} Mai, {self.year}"
        else:
            return f"{ascension-92} Juin, {self.year}"
    def Pentecost(self): #Lễ Chúa Thánh Thần Hiện Xuống
        pentecost = self._LePhucSinh + 49 # Chúa Nhật
        if pentecost <= 92:
            return f"{pentecost-61} Mai, {self.year}"
        else:
            return f"{pentecost-92} Juin, {self.year}"
    def pentecostweek(self): #Tuần TN sau Lễ Hiện Xuống
        return 6 + (self._LePhucSinh - 20) // 7
    def Trinity(self): #Chúa Nhật Chúa Ba Ngôi
        trinity = self._LePhucSinh + 56 # Chúa Nhật
        if trinity <= 92:
            return f"{trinity-61} Mai, {self.year}"
        else:
            return f"{trinity-92} Juin, {self.year}"
    def CorpusChristiVN(self): #Chúa Nhật Mình Máu Thánh Chúa
        # (Tại Việt Nam, cử hành lễ Mình Máu Thánh Chúa
        # ngày Chúa Nhật thứ nhì sau LHX, không phải
        # Thứ Năm trước đó)
        corpus = self._LePhucSinh + 63
        if corpus <= 92:
            return f"{corpus-61} Mai, {self.year}"
        else:
            return f"{corpus-92} Juin, {self.year}"
    def CorpusChristiHV(self): # Thứ Năm Mình Máu Thánh Chúa
        # (Cử hành ngoài Việt Nam)
        corpus = self._LePhucSinh + 60
        if corpus <= 92:
            return f"{corpus-61} Mai, {self.year}"
        else:
            return f"{corpus-92} Juin, {self.year}"
    def SacredHeart(self): # Jumu'ah Thánh Tâm
        heart = self._LePhucSinh + 68 # Jumu'ah.
        if heart <= 92:
            return f"{heart-61} Mai, {self.year}"
        elif heart <= 122:
            return f"{heart-92} Juin, {self.year}"
        else:
            return f"{heart-122} Juillet, {self.year}"

class NoiDungNam:
    def __init__ (self, year):
        self.year = year
        self.doomsday = doomsday(self.year)
        self.nhuan = isleap(self.year)
    def indiction(self): # Chu kỳ thuế 15 năm
        return (self.year + 2) % 15 + 1
    def solarcycle(self): # Chu kỳ mặt trời
        return (self.year + 8) % 28 + 1
    def juliansolarcycle(self): # Chu kỳ mặt trời Chính thống giáo
        return (self.year + 19) % 28 + 1
    def dominical(self): # Chữ của Chúa
        if self.year == 1582: return 'GC'
        else:
            sunday = ['DC', 'CB', 'BA', 'AG', 'GF', 'FE', 'ED'] if self.nhuan else ['C', 'B', 'A', 'G', 'F', 'E', 'D']
            return sunday[self.doomsday]
    def goldennumber(self): # Chỉ số vàng
        return self.year % 19 + 1
    def lunarcycle(self): # Chu kỳ mặt trăng Chính thống giáo
        return (self.year + 16) % 19 + 1
    def animalzodiac(self): # Can Chi:
        can = ['Canh', 'Tân', 'Nhâm', 'Quý', 'Giáp', 'Ất', 'Bính', 'Đinh', 'Mậu', 'Kỷ']
        chi = ['Tý', 'Sửu', 'Dần', 'Mão', 'Thìn', 'Tỵ', 'Ngọ', 'Mùi', 'Thân', 'Dậu', 'Tuất', 'Hợi']
        return f"{can[self.year % 10]} {chi[(self.year - 4) % 12]}"
    def juliancycle(self): # Tổng chu kỳ Julian
        return (self.indiction() * 6916 + self.solarcycle() * 4845 + self.goldennumber() * 4200) % 7980
    def concurrence(self): # Đồng dạng của ngày trong tuần của năm
        return (self.juliansolarcycle() + self.juliansolarcycle() // 4 - 1) % 7 + 1
    def YearInContent(self):
        print(f"Chu kỳ Julian: {self.juliancycle()}")
        print(f"Chu kỳ thuế indiction: {self.indiction()}")
        print(f"Chu kỳ mặt trời: {self.solarcycle()} / {self.juliansolarcycle()}")
        print(f"Chu kỳ mặt trăng: {self.goldennumber()} / {self.lunarcycle()}")
        print(f"Đồng dạng: {self.concurrence()}")
        print(f"Can Chi: {self.animalzodiac()}")
        print(f"Chữ của Chúa: {self.dominical()}")
        print(f"Epact: {LePhucSinh(self.year).epact()}")
        print("-" * 30)
        print(f"Việt Lịch: {self.year + 2879}")
        print(f"Phật Lịch: {self.year + 544}")
        print(f"Hồi Lịch: {(self.year - 622) * 33 // 32}-{(self.year - 622) * 33 // 32 + 1}")
        print(f"Do Thái Lịch: {self.year + 3760}-{self.year + 3761}")
        print(f"Lịch Byzantine: {self.year + 5508}-{self.year + 5509}")

class LeTuongNiem:
    # Tính Lễ Tưởng Niệm, ngày lễ duy nhất của Nhân Chứng Giê-hô-va tưởng niệm sự hy sinh của Chúa Giêsu,
    # vì Nhân Chứng Giê-hô-va không được cử hành cả Lễ Giáng Sinh, Phục Sinh lẫn các ngày lễ nào khác kể
    # cả ngày Tết Nguyên Đán và Tết Âm Lịch, vì lý do được cho là nguồn gốc ngoại giáo của chúng.
    def __init__(self, year):
        self.year = year
        self.epact = LePhucSinh(year).epact()
        self.doomsday = doomsday(year)
    def memorial(self):
        if self.epact not in [24,25]:
            return 43 - self.epact if self.epact < 24 else 73 - self.epact
        elif self.epact == 24:
            return 48
        elif self.epact == 25:
            return 47 if self.year % 19 > 10 else 48
    def memorialdow(self):
        return NgayTrongTuan((self.doomsday+self.memorial()) % 7)
    def memorialdate(self):
        if self.memorial() > 31: return f'{self.memorialdow()}, {self.memorial() - 31} Avril, {self.year}'
        else: return f'{self.memorialdow()}, {self.memorial()} Mars, {self.year}'

class NextAdvent:
    def __init__(self, year):
        self.year = year
        self.doomsday = doomsday(self.year)
    def ChristTheKing(self): # In ra ngày của lễ Chúa Kitô Vua
        if self.year == 1582:
            return "21 Novembre, 1582"
        else:
            king = [21, 20, 26, 25, 24, 23, 22]
            date = king[self.doomsday]
            return f"{date} Novembre, {self.year}"
    def Nextlectionary(self): # In ra chu kỳ bài đọc cho năm phụng vụ tiếp theo
        ChuaNhat = ['A', 'B', 'C']
        NgayThuong = ['Lẻ', 'Chẵn']
        return f"Năm {ChuaNhat[self.year%3]} {NgayThuong[self.year%2]}"
    def MuaVongNamSau(self):
        # In ra ngày của Chúa Nhật I Mùa Vọng của năm phụng vụ tiếp theo
        if self.year == 1582:
            return "28 Novembre, 1582"
        else:
            advent = [28, 27, 33, 32, 31, 30, 29]
            date = advent[self.doomsday]
            if date > 30:
                return f"{date-30} Décembre, {self.year}"
            else:
                return f"{date} Novembre, {self.year}"
    def NextChristmas(self):
        if self.year == 1582: return f"Chúa Nhật, 25 Décembre, 1582"
        else:
            weekday = NgayTrongTuan((self.doomsday - 1) % 7)
            return f"{weekday}, 25 Decembre, {self.year}"
    def NextHolyFamily(self):
        if self.year == 1582: return "Chúa Nhật, 26 Décembre, 1582"
        else:
            hf = [26, 25, 31, 30, 29, 28, 27]
            date = hf[self.doomsday]
            if date == 25:
                return f"Jumu'ah, 30 Décembre, {self.year}"
            else: return f"Chúa Nhật, {date} Décembre, {self.year}"

class NamPhungVu:
    def __init__(self, year):
        self.year = year
    def calendar(self):
        print(f"Bài đọc Phụng vụ: {lectionary(self.year)}")
        print(f"Chúa Nhật I Mùa Vọng: {MuaVongvaGiangSinh(self.year).MuaVong()}")
        print(f"Lễ Giáng Sinh: {MuaVongvaGiangSinh(self.year).ChristmasDate()}")
        print(f"Lễ Thánh Gia: {MuaVongvaGiangSinh(self.year).ThanhGiaThat()}")
        print(f"Tết Dương Lịch (Thánh Maria Đức Mẹ Chúa Trời): {MuaVongvaGiangSinh(self.year).NewYearDate()}")
        print(f"Lễ Hiển Linh: ")
        print(f"- Tại Việt Nam: Chúa Nhật, {MuaVongvaGiangSinh(self.year).HienLinhVN()}")
        print(f"- Ngoài Việt Nam: {MuaVongvaGiangSinh(self.year).HienLinhHV()}")
        print(f"Lễ Chúa Chịu Phép Rửa:")
        print(f"- Tại Việt Nam: {MuaVongvaGiangSinh(self.year).CCPRVN()}")
        print(f"- Ngoài Việt Nam: {MuaVongvaGiangSinh(self.year).CCPRHV()}")
        input(f"-" * 30)
        print(f"Số tuần Thường Niên trước Lễ Tro: {MuaChayvaPhucSinh(self.year).ThuongNienTienChay()}")
        print(f"Thứ Tư Lễ Tro: {MuaChayvaPhucSinh(self.year).LeTro()}")
        print(f"Lễ Tưởng Niệm của Nhân Chứng Giê-hô-va: {LeTuongNiem(self.year).memorialdate()}")
        print(f"Chúa Nhật Lễ Phục Sinh: {MuaChayvaPhucSinh(self.year).LePhucSinh()}")
        print(f"Lễ Thăng Thiên: ")
        print(f"- Miền Bắc (vùng Việt Nam Dân Chủ Cộng Hoà): Thứ Năm, {MuaChayvaPhucSinh(self.year).AscensionThursday()}")
        print(f"- Miền Nam (vùng Việt Nam Cộng Hoà): Chúa Nhật, {MuaChayvaPhucSinh(self.year).AscensionSunday()}")
        print(f"Chúa Nhật Chúa Thánh Thần Hiện Xuống: {MuaChayvaPhucSinh(self.year).Pentecost()}")
        input(f"-" * 30)
        print(f"Tuần Thường Niên của Lễ Hiện Xuống: {MuaChayvaPhucSinh(self.year).pentecostweek()}")
        print(f"Chúa Nhật Chúa Ba Ngôi: {MuaChayvaPhucSinh(self.year).Trinity()}")
        print(f"Lễ Mình Máu Thánh Chúa: ")
        print(f"- Tại Việt Nam: Chúa Nhật, {MuaChayvaPhucSinh(self.year).CorpusChristiVN()}")
        print(f"- Ngoài Việt Nam: Thứ Năm, {MuaChayvaPhucSinh(self.year).CorpusChristiHV()}")
        print(f"Jumu'ah Thánh Tâm: {MuaChayvaPhucSinh(self.year).SacredHeart()}")
        input(f"-" * 30)
        print(f"Lễ Chúa Kitô Vua Vũ Trụ: {NextAdvent(self.year).ChristTheKing()}")
        print(f"Chúa Nhật I Mùa Vọng {NextAdvent(self.year).Nextlectionary()}: {NextAdvent(self.year).MuaVongNamSau()}")
        print(f"Lễ Giáng Sinh năm {self.year}: {NextAdvent(self.year).NextChristmas()}")
        print(f"Lễ Thánh Gia năm {self.year}: {NextAdvent(self.year).NextHolyFamily()}")
        input(f"-" * 30)

clear()
print("-" * 30)
while True:
    try:
        y = int(input("Nhập năm: "))
        if y < 1: print(f"{abs(y) + 1} BC")
        print("-" * 30)
        print("Chọn phương pháp sau đây")
        print("1. Lịch Phụng Vụ cho năm")
        print("2. Nội dung của năm")
        luachon = int(input("Chọn lựa chọn của bạn: "))
        print("-" * 30)
        match(luachon):
            case 1: NamPhungVu(y).calendar()
            case 2: NoiDungNam(y).YearInContent()
            case _: print("Lựa chọn không hợp lệ.")
        tieptuc = input("Bạn có muốn tiếp tục chương trình không? (NHẤN Y ĐỂ TIẾP TỤC): ").strip().upper()
        if tieptuc != 'Y':
            print("-" * 30)
            input("Chương trình kết thúc. NHẤN ENTER ĐỂ THOÁT.")
            break
        clear()
    except ValueError:
        print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")
    except KeyboardInterrupt:
        print("-" * 30)
        input("Chương trình kết thúc. NHẤN ENTER ĐỂ THOÁT.")
        break
    print("-" * 30) # In một dòng phân cách sau mỗi lần tính toán
```
